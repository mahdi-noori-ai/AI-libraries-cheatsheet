# Seaborn Cheatsheet

## Import Libraries

```python

import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Load Dataset

```python

# Load example dataset
df = sns.load_dataset('iris')
print(df.head())
```

## Basic Scatter Plot

```python

# Scatter plot
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species')
plt.title('Scatter Plot of Sepal Length vs Sepal Width')
plt.show()
```

## Line Plot

```python

# Line plot
fmri = sns.load_dataset('fmri')
sns.lineplot(data=fmri, x='timepoint', y='signal', hue='region', style='event')
plt.title('Line Plot of FMRI Signal')
plt.show()
```

## Bar Plot

```python

# Bar plot
titanic = sns.load_dataset('titanic')
sns.barplot(x='class', y='survived', hue='sex', data=titanic)
plt.title('Bar Plot of Titanic Survival by Class and Sex')
plt.show()
```

## Histogram

```python

# Histogram
sns.histplot(data=df, x='petal_length', bins=20, kde=True)
plt.title('Histogram of Petal Length')
plt.show()
```

## Box Plot

```python

# Box plot
sns.boxplot(x='species', y='petal_length', data=df)
plt.title('Box Plot of Petal Length by Species')
plt.show()
```

## Violin Plot

```python

# Violin plot
sns.violinplot(x='species', y='petal_length', data=df)
plt.title('Violin Plot of Petal Length by Species')
plt.show()
```

## Pair Plot

```python

# Pair plot
sns.pairplot(df, hue='species')
plt.suptitle('Pair Plot of Iris Dataset', y=1.02)
plt.show()
```

## Joint Plot

```python

# Joint plot
sns.jointplot(data=df, x='sepal_length', y='sepal_width', hue='species', kind='kde')
plt.suptitle('Joint Plot of Sepal Dimensions', y=1.02)
plt.show()
```

## Regression Plot

```python

# Regression plot
sns.lmplot(data=df, x='petal_length', y='petal_width', hue='species', height=6)
plt.title('Regression Plot of Petal Dimensions')
plt.show()
```

## Facet Grid

```python

# Facet grid
g = sns.FacetGrid(titanic, col='survived', row='class', margin_titles=True)
g.map(sns.histplot, 'age')
plt.suptitle('Facet Grid of Titanic Age Distribution by Survival and Class', y=1.02)
plt.show()
```
